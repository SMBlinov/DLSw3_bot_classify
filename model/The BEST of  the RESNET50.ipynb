формируем датасет: проверочный сет будет составлять 20% от общего, из трансформаций оставляем только изменение яркости картинки в пределах 30% (флипы и повороты негативно сказываются на качество обучения..почему то). размер батча 64, размер картинки 299. Все это дело нормализуем.

```python
from pathlib import Path
PATH = Path('/content/dataset/images')
data = ImageDataBunch.from_folder(PATH,valid_pct=0.2, ds_tfms=get_transforms(do_flip=False,flip_vert=False, max_rotate=0,max_lighting=0.3),bs=64, size=299, num_workers=0).normalize(imagenet_stats)
```

Посмотрим что у нас в батче (выводит случайные картинки плиткой 6х6)

```python
data.show_batch(rows=3, figsize=(6,6))
```

Загружаем предобученную модель RESNET50

```python
model = cnn_learner(data, models.resnet50, metrics=accuracy)
```

Запускаем обучение на 3 эпохах (пока без параметров)

```python
model.fit_one_cycle(3)
```

Размораживаем слои и пытаемся подобрать learning rate


```python
# Здесь мы используем lr_finder, чтобы выбрать скорость обучения
model.unfreeze()
model.lr_find()
model.recorder.plot(suggestion=True)
```

Подставляем найденный lr и запускаем обучение

```python
# используем 1cycle policy подрбнее https://docs.fast.ai/callbacks.one_cycle.html
model.fit_one_cycle(10, 1.10e-06)
```

Посмотрим насколько качественно модель определяет картинки и на каких чащще всего ошибается. 

```python
interp = ClassificationInterpretation.from_learner(model)
interp.plot_confusion_matrix(figsize=(15,15), dpi=60)
```

```python
interp.plot_top_losses(5)
```
